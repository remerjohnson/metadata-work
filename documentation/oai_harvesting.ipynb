# OAI-PMH Harvesting 
This notebook will walk through OAI metadata harvesting using the python module `sickle` [Documentation](https://sickle.readthedocs.io/en/latest/index.html).

First, we import `sickle`:

```python
from sickle import Sickle
```

Next, we initialize the Sickle object and point it to the Data Science Journal OAI endpoint

```python
sickle = Sickle('https://datascience.codata.org/jms/index.php/up/oai')
```

Now, we can get all the records using the `.listRecords()` method

```python
records = sickle.ListRecords(metadataPrefix='oai_dc')
```

Now, we can see the raw XML records to make sure they were harvested

```python
for record in records:
    print(record)
```
